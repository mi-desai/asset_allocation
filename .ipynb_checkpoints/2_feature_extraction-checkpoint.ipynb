```python
import utilities as ut
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = ut.clean_data()
```

```python
test_data_1 = data[0]
test_data_2 = data[9]

test = [test_data_1, test_data_2]
```

```python
def extract_features(data):
    
    result = []
    trading_days = 252
    
    for i in enumerate(data):
        
        ts = i[1]
        ts.sort_index()
        colname = ts.columns[0]
        
        # lags
        ts['Lag_1year'] = ts[colname].shift(-1 * trading_days)
        ts['Lag_3year'] = ts[colname].shift(-1 * (trading_days * 3))
        ts['Lag_5year'] = ts[colname].shift(-1 * (trading_days * 5))
        ts['Lag_3month'] = ts[colname].shift(-1 * round(trading_days * 0.25))
        ts['Lag_6month'] = ts[colname].shift(-1 * round(trading_days * 0.5))
        ts['Lag_9month'] = ts[colname].shift(-1 * round(trading_days * 0.75))
        
        # rolling returns at various lags
        # PROBLEM RIGHT NOW THIS IS JUST ROLLING ASSET VALUE AVERAGES, NOT RETURNS
        ts['SMA_1year'] = ts[colname].rolling(window=str(trading_days) + 'D').mean()
        ts['SMA_3year'] = ts[colname].rolling(window=str(trading_days * 3) + 'D').mean()
        ts['SMA_5year'] = ts[colname].rolling(window=str(trading_days * 5) + 'D').mean()
        ts['SMA_3month'] = ts[colname].rolling(window=str(round(trading_days * 0.25)) + 'D').mean()
        ts['SMA_6month'] = ts[colname].rolling(window=str(round(trading_days * 0.5)) + 'D').mean()
        ts['SMA_9month'] = ts[colname].rolling(window=str(round(trading_days * 0.75)) + 'D').mean()
        
        # cumulative returns
        # need to look this up
        
        # rolling risk measurements - again - this is the standard deviation of the overall asset change / volatility of the asset value, assuming normal distribution
        ts['STD_1year'] = ts[colname].rolling(window=str(trading_days) + 'D').std()
        
        ts[[colname, 'SMA_6month', 'SMA_3year', 'SMA_1year']].loc['2018':'2020'].plot(figsize=(10, 5))
        plt.show()
        
        print(ts)
    
    
    
extract_features(test)
```
